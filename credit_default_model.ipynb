```python
from ucimlrepo import fetch_ucirepo 
# fetch dataset 
default_of_credit_card_clients = fetch_ucirepo(id=350) 

# data (as pandas dataframes) 
X = default_of_credit_card_clients.data.features 
y = default_of_credit_card_clients.data.targets 

# metadata 
print(default_of_credit_card_clients.metadata) 

# variable information 
print(default_of_credit_card_clients.variables) 

```

```python
import pandas as pd

# Combine X (features) and y (target) into one table
df = pd.concat([X, y], axis=1)

# Make empty lists for categorical and numerical columns
categorical = []
numerical = []

# Check each column in X
for col in X.columns:
    # If the column has 10 or fewer unique values, add to categorical
    if X[col].nunique() <= 10:
        categorical.append(col)
    # If more, add to numerical
    else:
        numerical.append(col)

# Show which columns are categorical and which are numerical        
print('Categorical:', categorical)
print('Numerical:', numerical)
```

```python
# For each categorical column, show how many times each value appears
for col in categorical:
    print(X[col].value_counts())
    
# For numerical columns, show basic statistics    
X[numerical].describe()
```

```python
# Copy DataFrame to save the original
df_labeled = df.copy()

# Decode categorical variables
df_labeled['X2'] = df_labeled['X2'].replace({1: 'Male', 2: 'Female'})
df_labeled['X3'] = df_labeled['X3'].replace({
    1: 'Graduate school',
    2: 'University',
    3: 'High school',
    4: 'Other',
    5: 'Other',
    6: 'Other',
    0: 'Unknown'
})
df_labeled['X4'] = df_labeled['X4'].replace({
    1: 'Married',
    2: 'Single',
    3: 'Other',
    0: 'Unknown'
})
df_labeled['Y'] = df_labeled['Y'].replace({0: 'No Default', 1: 'Default'})
df_labeled.head()
```

```python
# Add the target variable to the dataset
df = pd.concat([X, y], axis=1)

# Check how education is distributed across target classes
print("Education level by default status:")
print(df.groupby('Y')['X3'].value_counts(normalize=True).unstack().round(2))

# Marital status
print("\nMarital status by default status:")
print(df.groupby('Y')['X4'].value_counts(normalize=True).unstack().round(2))

# Gender
print("\nGender by default status:")
print(df.groupby('Y')['X2'].value_counts(normalize=True).unstack().round(2))
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create the encoder for categorical columns
encoder = OneHotEncoder(drop='first', sparse_output=False)

# Apply the encoder to categorical columns
X_cat_encoded = encoder.fit_transform(X[categorical])

# Convert the result to a DataFrame with column names
X_cat_encoded = pd.DataFrame(X_cat_encoded, columns=encoder.get_feature_names_out(categorical))
```

```python
from sklearn.preprocessing import StandardScaler

# Create the scaler for numerical columns
scaler = StandardScaler()

# Apply the scaler to numerical columns (make values have mean 0 and std 1)
X_num_scaled=scaler.fit_transform(X[numerical])

# Convert the result to a DataFrame with the original column names
X_num_scaled=pd.DataFrame(X_num_scaled, columns=numerical)
```

```python
# Combine the scaled numerical data and encoded categorical data into one table
X_final = pd.concat([X_num_scaled, X_cat_encoded], axis=1)
```

```python
from sklearn.model_selection import train_test_split 

# Split the data into training and test sets
# 80% goes to training, 20% goes to testing
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression 

# Create the logistic regression model with max 1000 iterations
logreg_model = LogisticRegression(max_iter=1000)

# Train the model on the training data
logreg_model.fit(X_train, y_train.values.ravel())
```

```python
# Use the trained model to make predictions on the test data
y_pred_logreg = logreg_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Show the confusion and detailed matrix
print(confusion_matrix(y_test, y_pred_logreg))
print(classification_report(y_test, y_pred_logreg))
```

```python
from sklearn.ensemble import RandomForestClassifier 

# Create the Random Forest model with 100 trees
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42) 

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train.values.ravel())

y_pred_rf = rf_model.predict(X_test)

print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
mlp.fit(X_train, y_train.values.ravel())

y_pred_mlp = mlp.predict(X_test)

print("Results:")
print(confusion_matrix(y_test, y_pred_mlp))
print(classification_report(y_test, y_pred_mlp))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns

#Define metrics function
def get_metrics(y_true, y_pred, model_name): 
    return {
        'Model': model_name, 
        'Accuracy': accuracy_score(y_true, y_pred), 
        'Precision': precision_score(y_true, y_pred),
        'Recall': recall_score(y_true, y_pred),
        'F1-score': f1_score(y_true, y_pred)
    }

#Calculate metrics
metrics_logreg = get_metrics(y_test, y_pred_logreg, 'LogisticRegression')
metrics_rf = get_metrics(y_test, y_pred_rf, 'RandomForest')
metrics_mlp = get_metrics(y_test, y_pred_mlp, 'MLPClassifier')

#Create DataFrame
results_df = pd.DataFrame([metrics_logreg, metrics_rf, metrics_mlp])

#Print table
print("\n=== Model Metrics Table ===")
print(results_df.to_string(index=False))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

results_melted = results_df.melt(id_vars='Model', var_name='Metric', value_name='Score')

plt.figure(figsize=(10, 6))
sns.lineplot(
    data=results_melted,
    x='Metric',
    y='Score',
    hue='Model',
    marker='o'
)

plt.title('Scatter Plot of Model Metrics')
plt.ylim(0, 1)
plt.grid(True)
plt.legend(title='Model')
plt.show()
```
